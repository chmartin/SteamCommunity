```python
playtime_per_user = pd.DataFrame(df.groupby('steamid').sum()[['playtime_forever']])
```

```python
count_per_user = df.groupby('steamid').count()
```

```python
mean_per_user = df.groupby('steamid').mean()
```

```python
std_per_user = df.groupby('steamid').std()
```

```python
max_per_user = df.loc[df.groupby(["steamid"])["playtime_forever"].idxmax()]
```

```python
playtime_per_user = playtime_per_user.reset_index()
count_per_user = count_per_user.reset_index()
mean_per_user = mean_per_user.reset_index()
std_per_user = std_per_user.reset_index()
```

```python
print(playtime_per_user.info())
print(count_per_user.info())
print(mean_per_user.info())
print(std_per_user.info())
print(max_per_user.info())
```

```python
total_per_user = playtime_per_user.merge(count_per_user,on='steamid',suffixes=('_sum', '_count'))
```

```python
total_per_user = total_per_user.merge(mean_per_user,on='steamid',suffixes=('_count','_mean'))
```

```python
total_per_user = total_per_user.merge(std_per_user,on='steamid',suffixes=('_mean','_std'))
```

```python
total_per_user = total_per_user.merge(max_per_user,on='steamid',suffixes=('_std','_max'))
```

```python
total_per_user.head()
```

```python
total_per_user = total_per_user.drop(['playtime_2weeks_count','appid_mean','playtime_2weeks_mean','appid_std','playtime_2weeks_std','playtime_2weeks_max'],axis='columns')
```

```python
total_per_user[total_per_user['playtime_forever_count'] != total_per_user['appid_count']]
```

```python
total_per_user = total_per_user.drop('playtime_forever_count',axis='columns')
```

```python
total_per_user['playtime_forever_max'] = total_per_user['playtime_forever']
```

```python
total_per_user = total_per_user.drop('playtime_forever',axis='columns')
```

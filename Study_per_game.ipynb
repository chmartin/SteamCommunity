```python
playtime_per_game = pd.DataFrame(df.groupby('appid').sum()[['playtime_forever']])
```

```python
count_per_game = df.groupby('appid').count()
```

```python
mean_per_game = df.groupby('appid').mean()
```

```python
std_per_game = df.groupby('appid').std()
```

```python
playtime_per_game = playtime_per_game.reset_index()
count_per_game = count_per_game.reset_index()
mean_per_game = mean_per_game.reset_index()
std_per_game = std_per_game.reset_index()
```

```python
print(playtime_per_game.head())
print(count_per_game.head())
print(mean_per_game.head())
print(std_per_game.head())
```

```python
total_per_game = playtime_per_game.merge(count_per_game,on='appid',suffixes=('_sum', '_count'))
```

```python
total_per_game = total_per_game.merge(mean_per_game,on='appid',suffixes=('_count','_mean'))
```

```python
total_per_game = total_per_game.merge(std_per_game,on='appid',suffixes=('_mean','_std'))
```

```python
total_per_game = total_per_game.drop(['steamid','playtime_2weeks','steamid_mean','playtime_2weeks_mean','playtime_2weeks_count'],axis='columns')
```

```python
total_per_game[total_per_game['playtime_forever_count'] != total_per_game['steamid_count']]
```

```python
total_per_game = total_per_game.drop('steamid_count',axis='columns')
```
